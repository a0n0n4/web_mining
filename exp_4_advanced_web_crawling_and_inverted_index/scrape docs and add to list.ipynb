```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import textract
```

```python
l=[]
textPage = urlopen('http://www.pythonscraping.com/pages/warandpeace/chapter1.txt')
l.append(textPage.read())

textPage = urlopen('http://www.pythonscraping.com/pages/warandpeace/chapter1-ru.txt')
l.append(str(textPage.read(), 'utf-8'))

URL = "http://pythonscraping.com/pages/AWordDocument.docx"
response = requests.get(URL)
open("AWordDocument.docx", "wb").write(response.content)
text = textract.process("./AWordDocument.docx")
l.append(text)

URL = "http://pythonscraping.com/pages/warandpeace/chapter1.pdf"
response = requests.get(URL)
open("APdfDocument.pdf", "wb").write(response.content)
text = textract.process("./APdfDocument.pdf")
l.append(text)
```
